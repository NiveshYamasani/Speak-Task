### Cleaning salary variable

```python
for i in range(len(max_sal)):
    if '₹' in max_sal[i] and '₹' in min_sal[i]:
        max_sal[i]=max_sal[i].replace('₹','')
        min_sal[i]=min_sal[i].replace('₹','') 
    max_sal[i]=max_sal[i].replace(',','')
    max_sal[i]=max_sal[i].replace('a year','').strip()
    min_sal[i]=min_sal[i].replace(',','')
    min_sal[i]=min_sal[i].replace('a year','').strip()
```

```python
min_sal
```

```python
max_sal
```

```python
for i in range(len(min_sal)):
    if 'a month' in min_sal[i]:
        min_sal[i]=min_sal[i].replace('a month','').strip()
        s=int(min_sal[i].strip())
        s=s*12
        min_sal[i]=s
    elif 'an hour' in min_sal[i]:
        min_sal[i]=min_sal[i].replace('an hour','').strip()
        min_sal[i]=40*52*int(min_sal[i])
    elif 'a week' in min_sal[i]:
        min_sal[i]=min_sal[i].replace('a week','').strip()
        min_sal[i]=52*int(min_sal[i])
    elif 'a day' in min_sal[i]:
        min_sal[i]=min_sal[i].replace('a day','').strip()
        min_sal[i]=365*int(min_sal[i])
```

```python
for i in range(len(max_sal)):
    if 'a month' in max_sal[i]:
        max_sal[i]=max_sal[i].replace('a month','').strip()
        s=int(max_sal[i].replace(',','').strip())
        s=s*12
        max_sal[i]=s
    elif 'an hour' in max_sal[i]:
        max_sal[i]=max_sal[i].replace('an hour','').strip()
        max_sal[i]=40*52*int(max_sal[i])
    elif 'a week' in max_sal[i]:
        max_sal[i]=max_sal[i].replace('a week','').strip()
        max_sal[i]=52*int(max_sal[i])
    elif 'a day' in max_sal[i]:
        max_sal[i]=max_sal[i].replace('a day','').strip()
        max_sal[i]=365*int(max_sal[i])
    
    
```

```python
df['Minimum_salary']=min_sal
df['Maximum_salary']=max_sal
```

```python
def job_title(x):
    if 'Software Developer' in x or 'Software Engineer' in x or 'Developer' in x or 'developer' in x :
        return 'Software Developer'
    elif 'Quality' in x or 'Test' in x or'Test Engineer' in x or 'Quality Engineer' in x or 'Automation' in x or 'Engineer' in x:
        return 'Testing Engineer'
    elif 'accountant' in x or 'Accountant' in x or 'Senior Accountant' in x or 'Sr. Accountant' in x or 'ACCOUNT' in x or '':
        return 'Accountant'
    elif 'Relationship Manager' in x or 'Manager' in x or 'manager' in x or 'HR Executive' in x or 'HR' in x or 'MANAGER' in x:
        return 'Hr Manager'
    elif 'Data Science' in x or 'Data' in x or 'data' in x or 'Machine Learning' in x or 'NLP' in x:
        return 'Data Scientist'
    else:
        return 'others'
    
```

```python
df['job_title']=df['job_title'].apply(job_title)
df.head()
```

```python
def location(x):
    if 'Telangana' in x or 'Hyderabad' in x :
        return 'Hyderabad'
    elif 'Chennai' in x or 'Tamilnadu' in x:
        return 'Chennai'
    elif 'Maharastra' in x or 'Mumbai' in x:
        return 'Mumbai'
    elif 'Bangalore' in x or 'Karnataka' in x:
        return 'Bangalore'
    elif 'Delhi' in x:
        return 'Delhi'
    else:
        return 'others'
```
